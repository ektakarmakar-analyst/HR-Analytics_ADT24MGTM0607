```python
sns.countplot(x='Attrition', data=df)
plt.title('Distribution of Attrition')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
attrition_counts=df['Attrition'].value_counts()
total_employees=attrition_counts.sum()
attrition_percentages=(attrition_counts/total_employees)*100
print(attrition_percentages)
print(attrition_counts)
```

```python
attrition_counts=df['Attrition'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(attrition_counts,labels=attrition_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Distribution of Attrition')
plt.show()
```

```python
department_counts=df['Department'].value_counts()
total_employees=department_counts.sum()
department_percentages=(department_counts/total_employees)*100
print(department_percentages)
print(department_counts)
```

```python
department_counts=df['Department'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(department_counts,labels=department_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Distribution of Department')
```

```python
jobrole_counts=df['JobRole'].value_counts()
total_employees=jobrole_counts.sum()
jobrole_percentages=(jobrole_counts/total_employees)*100
print(jobrole_percentages)
print(jobrole_counts)
```

```python
jobrole_counts=df['JobRole'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(jobrole_counts,labels=jobrole_counts.index,autopct='%1.1f%%',startangle=90)
```

```python
# Group by Department and calculate the mean of Attrition_numeric
attrition_by_department = df.groupby('Department')['Attrition_numeric'].mean().sort_values(ascending=False)

print("Attrition Rate by Department:")
print(attrition_by_department)
```

```python
attrition_by_department = df.groupby('Department')['Attrition_numeric'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=attrition_by_department.index, y=attrition_by_department.values)
```
